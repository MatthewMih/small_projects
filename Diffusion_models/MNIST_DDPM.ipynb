### Диффузионная моель для генерации изображений

(Используется метод, предложенный в статье DDPM (Jonathan Ho, Ajay Jain, Pieter Abbeel, 2020, https://arxiv.org/abs/2006.11239)

Модель для изображений 1x32x32 (Для примера обучалось на MNIST)

Для проверки генерации -- запустите все ячейки (кроме загрузки датасета, обучения и сохранения чекпоинта)

```python
import torch
import torchvision
import numpy as np
import random
import math

from torch import nn
from tqdm import trange
from matplotlib import pyplot as plt
```

## Загрузка датасета (optional):

```python
batchsize = 512 # уменьшить, если не хватает памяти видеокарты :))


transform = torchvision.transforms.Compose(
    [
        torchvision.transforms.Resize((32, 32)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(0.13, 0.285, inplace=True),
    ]
)

dataset = torchvision.datasets.MNIST('../../Datasets', transform=transform, download=True)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batchsize, shuffle=True)
```

## Классы, используемые для создания модели

Архитектура модели похожа на UNet, в котором обычные свертки заменены на MBConv-слои (подробнее о них -- в проекте "SHARE_research")

```python
class Parallel(nn.Module): # штучка для residual connection'ов, используем ее в MBConv
    def __init__(self, block1, block2): # ко входному тензору примерняются block1, block2 и складываются рез-таты. 
        super().__init__()
        self.block1 = block1
        self.block2 = block2
        
    def forward(self, x):
        return self.block1(x) + self.block2(x)
    

class MBConv(nn.Sequential):
    def __init__(self, in_c, out_c, inner_c=None):
        if inner_c == None:
            inner_c = 4 * in_c
        super().__init__(
            Parallel(
                nn.Sequential(
                    nn.Conv2d(in_c, inner_c, kernel_size=1),
                    nn.BatchNorm2d(inner_c),
                    nn.ReLU(),
                    
                    nn.Conv2d(inner_c, inner_c, kernel_size=3, groups=inner_c, padding=1),
                    nn.BatchNorm2d(inner_c),
                    nn.ReLU(),
                    
                    nn.Conv2d(inner_c, out_c, kernel_size=1),
                    nn.BatchNorm2d(out_c),
                    nn.ReLU(),
                ),
                nn.Conv2d(in_c, out_c, kernel_size=1)
            )
        )
        

class Left_side_block(nn.Module): # для UNet'a...
    def __init__(self, in_channels, out_channels):
        super().__init__()
        
        self.convs = nn.Sequential(
            MBConv(in_channels, out_channels),
            MBConv(out_channels, out_channels)
        )

    def forward(self, x):
        return self.convs(x)
    

class Right_side_block(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        
        self.upscale = nn.ConvTranspose2d(in_channels, in_channels, kernel_size=2, stride=2)
        
        self.convs = nn.Sequential(
            MBConv(2*in_channels, in_channels),
            MBConv(in_channels, in_channels // 2),
        )

    def forward(self, x_skip, x_small):
        x_upscaled = self.upscale(x_small)
        x = torch.cat((x_skip, x_upscaled), dim=1) # 2*in_channels
        return self.convs(x)
                      
# изображение in_channels x 32 x 32
class MicroUnet32_MB(nn.Module):
    def __init__(self, channels):
        super().__init__()
        
                      
        self.maxpool = torch.nn.MaxPool2d(2)
        self.downstep1 =  Left_side_block(in_channels=channels, out_channels=32)
        self.downstep2 =  Left_side_block(in_channels=32, out_channels=64)
        self.downstep3 =  Left_side_block(in_channels=64, out_channels=128)
                      
        self.middle = nn.Sequential(
            torch.nn.MaxPool2d(2),
                                    
            MBConv(128, 128),
            MBConv(128, 128)
        )
        
        self.upstep3 = Right_side_block(in_channels=128)
        self.upstep2 = Right_side_block(in_channels=64)
        self.upstep1 = Right_side_block(in_channels=32)
                      
        self.lastconv = nn.Sequential(            
            nn.Conv2d(16, channels, kernel_size=3, padding=1)
        )
                      

    def forward(self, x):
        x1 = self.downstep1(x) # out: 32x32x32
        x2 = self.downstep2(self.maxpool(x1)) # out: 64x16x16
        x3 = self.downstep3(self.maxpool(x2)) # out: 64x8x8
                      
        y3 = self.middle(x3) # out: 128x4x4
        
        y2 = self.upstep3(x3, y3) # out: 64x8x8
        y1 = self.upstep2(x2, y2) # out: 32x16x16
        y0 = self.upstep1(x1, y1) # out: 16x32x32
        return self.lastconv(y0)
```

## Функции для обучения модели и генерации изображений:

```python
def beta_func(t,T=100):
    return 1e-4 + (t - 1) / float(T - 1)*(1e-2 - 1e-4)
            

def alpha_func(t, T=100):
    return 1. - beta_func(t, T)

def alpha_bar_func(t, T=100):
    alpha_bar_result = 1.
    for k in range(t):
        alpha_bar_result *= alpha_func(k+1, T)
    return alpha_bar_result

def sigma_square_func(t, T=100):
    return (1. - alpha_bar_func(t-1, T)) / (1. - alpha_bar_func(t, T)) * beta_func(t, T)

def generate_images(model, device, n=64, T=100, noise=None):
    with torch.no_grad():
        shape = [n, 1, 32, 32]
        if noise == None:
            noise = torch.normal(torch.zeros([T]+shape, device=device),
                                 torch.ones([T]+shape, device=device)).to(device)
        x = noise[T-1][:n].clone().detach().to(device)
    
        for t in range(T, 0, -1):
            alpha_bar = alpha_bar_func(t, T)
            alpha = alpha_func(t, T)
        
            x -= (1. - alpha) / math.sqrt(1. - alpha_bar + 1e-7) * model(x)
            x /= (math.sqrt(alpha)+1e-7)
        
            if t != 1:
                sigma_square = sigma_square_func(t, T)
                x += math.sqrt(sigma_square)*noise[t-2]
    return x

def num_of_params(model, only_trainable: bool = False): # beautiful function from StackOverflow
    parameters = list(model.parameters())
    if only_trainable:
        parameters = [p for p in parameters if p.requires_grad]
    unique = {p.data_ptr(): p for p in parameters}.values()
    return sum(p.numel() for p in unique)

def plot_batch(x):
    grid = torchvision.utils.make_grid(x, scale_each=True, padding=1)
    plt.matshow(grid.permute(1, 2, 0).cpu().detach().numpy())
    plt.show()


def train(model, device, optimizer, dataloader, loss_list, epochs_n=100, T=1000, debug_mode=False, fixed_noise=None):
    loss_function = nn.MSELoss()
    for epoch in trange(epochs_n):
        accumulated_loss = 0
        
        model.train()
        for x, _ in dataloader:            
            t = random.randint(1, T)
            standard_noise = torch.normal(torch.zeros(x.shape, device=device),
                                          torch.ones(x.shape, device=device)).to(device)
            alpha_bar = alpha_bar_func(t, T)
            alpha = alpha_func(t, T)
            beta = beta_func(t, T)
            sigma_square = sigma_square_func(t, T)            
            
            x = x.to(device)
            x_noised = math.sqrt(alpha_bar)*x + math.sqrt(1. - alpha_bar)*standard_noise
            loss = loss_function(standard_noise, model(x_noised))
            loss *= (beta**2) / (2*sigma_square*alpha*(1-alpha_bar) + 1e-5)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            accumulated_loss += float(loss)
        model.eval()
        loss_list.append(accumulated_loss)
        
        if debug_mode:
            print("After ep ", len(loss_list), " loss: ", loss_list[-1], ": ")
            plot_batch(generate_images(model, device, n=16, T=T, noise=fixed_noise))
            plt.plot(loss_list)
            plt.show()
            print("==================")
            
            torch.save({'model_state_dict': model.state_dict(),
                        'optimizer_state_dict': optimizer.state_dict(),
                        'loss_list': loss_list},
                        str(len(loss_list))+"ep.pt")
```

## Инициализация

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Тензор шума, используемый для генерации изображений при обучении в debug режиме
fixed_noise = torch.normal(torch.zeros([1000, 16, 1, 32, 32], device=device),
                           torch.ones([1000, 16, 1, 32, 32], device=device)).to(device)

# Инициализация модели и оптимизатора:
model = MicroUnet32_MB(channels=1).to(device)
optimizer = torch.optim.Adam(model.parameters())
loss_list = []

print("Количество параметров модели: ", num_of_params(model))
```

## Обучение модели (optional):

```python
train(model, device, optimizer, dataloader, loss_list,
      epochs_n=50,T=1000, debug_mode=True, fixed_noise=fixed_noise)
```
